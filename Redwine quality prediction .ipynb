**Creating a function for remove outlier (IQR)**

```python
out=[]
def outliers_remove(df):
    q1 = df.quantile(0.25) # To find the 1st quantile
    q3 = df.quantile(0.75) # To find the 3rd quantile
    iqr = q3-q1 # To find the inter quantile range
    Lower_tail = q1 - 1.5 * iqr # To find the Lower fence
    Upper_tail = q3 + 1.5 * iqr # To find the Higher fence
    for i in df:
        if i > Upper_tail or i < Lower_tail:
            out.append(i)
            df.replace(i,np.nan, inplace=True)
            df.fillna(df.mean(), inplace=True)
    print("Outliers:",out)
```

```python
plt.title(' Distribution plot of Fixed Acidity')
sns.distplot(df['fixed acidity'])
```

--We can tell from the distribution plot that the values in this column are right skewed.

**`NOW REMOVING OUTLIER IN THIS COLUMN`**

```python
outliers_remove(df['fixed acidity'])
```
